```python
##Euclidean Distance
A = [1,1]
B = [2,2]
def Euclidean(A,B):
    distance = ((A[0]-B[0])**2 + (A[1]-B[1])**2)**(1/2)
    print(distance)
    return round(distance,2)
Euclidean(A,B)
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = {'A': [1, 1], 'B': [6, 6], 'C': [7, 7], 'D': [3, 2],
        'E': [6, 8], 'F': [2, 1], 'G': [2, 3]}
C1 = [4,3]
C2 = [7,9]
# Clear the current figure to avoid plotting on top of previous plots if this cell is run multiple times
plt.clf()

for name, coords in data.items():
    plt.plot(coords[0], coords[1], 'o', label=name) # Plot each point with a marker
    plt.text(coords[0] + 0.1, coords[1] + 0.1, name) # Add text label near each point

plt.plot(C1[0], C1[1], 'x', c='blue', markersize=8, label=f'Centroid C1') # Plot C1
plt.plot(C2[0], C2[1], 'x', c='red', markersize=8, label=f'Centroid C2') # Plot C2

# Add text labels for centroids
plt.text(C1[0] + 0.1, C1[1] + 0.1, f'C1')
plt.text(C2[0] + 0.1, C2[1] + 0.1, f'C2')

plt.xlabel('X-coordinate')
plt.ylabel('Y-coordinate')
plt.title('Plot of Points from Dictionary')
plt.grid(True)
plt.legend()
plt.show()
```

```python
C1 = [4,3]
C2 = [7,9]
d_C1=[]
d_C2=[]
for name, coords in data.items():
    d1= Euclidean(C1, coords)
    d2= Euclidean(C2, coords)
    d_C1.append(d1)
    d_C2.append(d2)

print(d_C1,d_C2)
```

```python
data = {'A': [1, 1], 'B': [6, 6], 'C': [7, 7], 'D': [3, 2],
        'E': [6, 8], 'F': [2, 1], 'G': [2, 3]}
```

```python
S1=[]
S2=[]
names = list(data.keys()) # Get the ordered list of keys (names)
for i in range(len(d_C1)):
    current_name = names[i] # Get the name corresponding to the current index
    if d_C1[i] > d_C2[i]:
        S1.append(current_name) # Append the name (key) to S1
    else:
        S2.append(current_name) # Append the name (key) to S2
print(S1, S2)
```

```python
S1_coords = [data[name] for name in S1]
C1 = np.mean(S1_coords, axis=0)

S2_coords = [data[name] for name in S2]
C2 = np.mean(S2_coords, axis=0)
print(C1, C2)
```

```python
for name, coords in data.items():
    plt.plot(coords[0], coords[1], 'o', label=name) # Plot each point with a marker
    plt.text(coords[0] + 0.1, coords[1] + 0.1, name) # Add text label near each point
plt.plot(C1[0], C1[1], 'x', c='blue', markersize=8, label=f'Centroid C1') # Plot C1
plt.plot(C2[0], C2[1], 'x', c='red', markersize=8, label=f'Centroid C2') # Plot C2

# Add text labels for centroids
plt.text(C1[0] + 0.1, C1[1] + 0.1, f'C1')
plt.text(C2[0] + 0.1, C2[1] + 0.1, f'C2')

plt.xlabel('X-coordinate')
plt.ylabel('Y-coordinate')
plt.title('Plot of Points from Dictionary with Centroids')
plt.grid(True)
plt.legend()
plt.show()
```
